```python
from matplotlib import pyplot as plt
```

```python
x_axis = [1, 2, 3, 4, 5, 6, 7, 8]
y_axis1 = [12300, 20000, 25000, 30000, 34000, 40000, 60000, 74000]
y_axis2 = [10000, 12000, 20000, 25000, 30000, 35000, 45000, 50000]
#to check different styles of plot avaiable
plt.style.available
```

```python
# plt.style.use('seaborn-talk')
plt.xkcd()  # Mimic comic style 
plt.plot(x_axis, y_axis1, color = 'k', linestyle = '--', marker = 'o', label = 'axis1')
plt.plot(x_axis, y_axis2, label = 'axis2')

plt.title('My first Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
# plt.legend(['axis1', 'axis2'])
plt.legend()
# plt.grid()
plt.show()
'''
two ways to add legends are either by giving an ordered list of labels and
***pass it to plt.legends()
   OR
   Add legends in plt.plot() itself using label = 'value'
***To distinguish plot lines, format is '[color][marker][line]'
   OR
   Add color = '', linestyle = '' in plt.plot() to make code more explicit
***To change thickness use linewidth = int in plt.plot() "default is 1"
'''
```
